```python
import pandas as pd
```

```python
group1 = pd.read_excel('../data/reflect_write/reflect_write_dmg.xlsx', sheet_name='A', header=None)
group2 = pd.read_excel('../data/reflect_write/reflect_write_dmg.xlsx', sheet_name='B', header=None)

# Set the third row as the header
new_header = group1.iloc[2] 
group1 = group1[3:]  # Remove the data above the third row including the third row
group1.columns = new_header  # Set the new header
new_header = group2.iloc[2] 
group2 = group2[3:]  # Remove the data above the third row including the third row
group2.columns = new_header  # Set the new header

# Reset the index of the DataFrame
group1.reset_index(drop=True, inplace=True)
group2.reset_index(drop=True, inplace=True)
```

```python
# "Hast du am Donnerstag in der Gruppe oder am Freitag in der Gruppe teilgenommen?" -> "Did you take part in the group on Thursday or the group on Friday?"
# 0 for Thursday, 1 for Friday
full = pd.concat([group1[['User', 'Donnerstag']], group2[['User', 'Donnerstag']]], axis=0)
full.rename(columns={'Donnerstag': 'Group'}, inplace=True)
full = full.drop_duplicates(subset='User', keep='first')
full
```
